```python
def parse_pos_int(s: str) -> int:
    n = int(s)
    if n < 0:
        raise ValueError(f"No negativity allowed, but got {n}")
    return n
```

```python
parse_pos_int("-5")
```

```python
parse_pos_int("a")
```

```python
%%ipytest
import pytest

def test_negative():
    with pytest.raises(ValueError, match=r"No negativity allowed"):
        parse_pos_int("-1")

def test_invalid():
    with pytest.raises(ValueError, match=r"invalid literal for int"):
        parse_pos_int("a")
```
